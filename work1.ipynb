pandasを利用してポケモンの名前、個体値、努力値、レベルから各ステータスの実数値を返すプログラムを作成


---



```python
#ポケモンsvの準伝説ポケモンの種族値を設定
import pandas as pd
df=pd.DataFrame([[155,80,85,55],[110,120,85,80],[125,80,100,80],[55,90,95,135],[80,65,135,120],[45,135,70,100]])
df.index=['H','A','B','C','D','S']
df.columns=['ディンルー','パオジアン','チオンジェン','イーユイ']
df
```

```python
#ポケモンの名前を入力
name=input('ポケモンの名前を入力してください:',)
#データフレームに情報があるか判定し、無ければ追加
if name not in df:
  print('そのポケモンは登録されていません\n新しく登録します')
  h=int(input('HP種族値を入力してください:',))
  a=int(input('攻撃種族値を入力してください:',))
  b=int(input('防御種族値を入力してください:',))
  c=int(input('特攻種族値を入力してください:',))
  d=int(input('特防種族値を入力してください:',))
  s=int(input('素早さ種族値を入力してください:',))
  df[name]=h,a,b,c,d,s
#レベルを入力
lv=int(input('レベルを入力してください:',))
#個体値をそれぞれ入力、値が適切か判定
while True:
  indh=int(input('HP個体値を入力してください（０～３１）:',))
  if 0 <= indh and indh <= 31:
    break
  print('個体値は０～３１の値にしてください\n')
while True:
  inda=int(input('攻撃個体値を入力してください（０～３１）:',))
  if 0 <= inda and inda <= 31:
    break
  print('個体値は０～３１の値にしてください\n')
while True:
  indb=int(input('防御個体値を入力してください（０～３１）:',))
  if 0 <= indb and indb <= 31:
    break
  print('個体値は０～３１の値にしてください\n')
while True:
  indc=int(input('特攻個体値を入力してください（０～３１）:',))
  if 0 <= indc and indc <= 31:
    break
  print('個体値は０～３１の値にしてください\n')
while True:
  indd=int(input('特防個体値を入力してください（０～３１）:',))
  if 0 <= indd and indd <= 31:
    break
  print('個体値は０～３１の値にしてください\n')
while True:
  inds=int(input('素早さ個体値を入力してください（０～３１）:',))
  if 0 <= inds and inds <= 31:
    break
  print('個体値は０～３１の値にしてください\n')
#努力値をそれぞれ入力、値が適切か判定
while True:
  while True:
    h=int(input('HPの努力値を入力してください（０～２５２）:',))
    if h < 0 or h > 252:
      print('努力値は０～２５２の値にしてください\n')
      continue
    break
  while True:
    a=int(input('攻撃の努力値を入力してください（０～２５２）:',))
    if a < 0 or a > 252:
        print('努力値は０～２５２の値にしてください\n')
        continue
    break
  while True:
    b=int(input('防御の努力値を入力してください（０～２５２）:',))
    if b < 0 or b > 252:
        print('努力値は０～２５２の値にしてください\n')
        continue
    break
  while True:
    c=int(input('特攻の努力値を入力してください（０～２５２）:',))
    if c < 0 or c > 252:
        print('努力値は０～２５２の値にしてください\n')
        continue
    break
  while True:
    d=int(input('特防の努力値を入力してください（０～２５２）:',))
    if d < 0 or d > 252:
        print('努力値は０～２５２の値にしてください\n')
        continue
    break
  while True:
    s=int(input('素早さの努力値を入力してください（０～２５２）:',))
    if s < 0 or s > 252:
        print('努力値は０～２５２の値にしてください\n')
        continue
    break
  sum=h+a+b+c+d+s
  if sum >= 0 or sum <= 510:
    break
  print('努力値合計は５１０までです\n')
#後のためにデータをリストにまとめる
data=[a,b,c,d,s]
data2=['A','B','C','D','S']
data3=[inda,indb,indc,indd,inds]
data4=[]
#実数値計算　HPとそれ以外で少し違う
HP=int((df[name]['H']*2+indh+h/4)*lv/100+60)
data4.append(HP)
for x in range(5):
  point=int((df[name][data2[x]]*2+data3[x]+data[x]/4)*lv/100+5)
  data4.append(point)
print(f'HP:{data4[0]},攻撃：{data4[1]},防御：{data4[2]},特攻：{data4[3]},特防：{data4[4]},素早さ：{data4[5]}')
```
